```python
import os
import glob
import re
import json

import pandas as pd
```

```python
calc_files = set(map(lambda x: x[:-6], glob.glob("*.txt")))
print("Calculated files", len(calc_files))
calc_files
```

```python
pattern = re.compile(r"{.+}")


rows = []
for filename in calc_files:
    batches = glob.glob("{}*".format(filename))
    for batch_file in batches:
        with open(batch_file) as f:
            last_line = f.readlines()[-1] 
            matches = pattern.findall(last_line)
            if len(matches) != 1:
                continue
            row = json.loads(matches[0])
            row['name'] = filename
        rows.append(row)
    
metrics_df = pd.DataFrame(rows)
metrics_df
```

```python
metrics_df.groupby("name").mean()
```
