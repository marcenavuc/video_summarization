```python
import os
import subprocess
import glob

import pandas as pd


ann_df = pd.read_parquet("../datasets/annotation.parquet")
videos = glob.glob("../datasets/videos/*.mp4")
print("All amount of videos", len(videos))
videos
```

```python
calc_files = set(map(lambda x: x[:-6], glob.glob("*.txt")))
print("Calculated files", len(calc_files))
calc_files
```

```python
for video_path in videos[:]:
    if os.path.basename(video_path)[:-4] in calc_files:
        videos.remove(video_path)
len(videos)
```

```python
cnt = 0
for video_path in videos:
    filename = os.path.basename(video_path).split(".")[0]
    cnt += ann_df[ann_df['video_id'] == filename].shape[0]

assert cnt == len(videos), "Some videos have not found"
```

```python
REPS = 5

for video_path in videos:
    for i in range(REPS):
        video_folder = video_path.split("/")[2]
        video_name = video_path.split("/")[3]
        res_name = video_name.split('.')[0]
        exec_str = "python ../video_summarization/pipelines/predict.py video_path={folder}/{name} > {res_name}_{rep}.txt 2>&1".format(
            folder=video_folder, name=video_name, res_name=res_name, rep=i
        )
        print(exec_str)
        os.system(exec_str)
```
